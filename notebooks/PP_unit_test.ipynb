```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the median filtered and ema cleaned data CSV file
clean_df = pd.read_csv('../data/Full_clean_SSTEMP_sensor.csv')

#Read the EMA cleaned data CSV file
ema_df = pd.read_csv('../data/ema_clean_SSTEMP_sensor.csv')

#Read the Median filtered data CSV file
med_df = pd.read_csv('../data/med_clean_SSTEMP_sensor.csv')

# Read the raw data CSV file
raw_df = pd.read_csv('../data/PP_TEST.csv')

#Format time colums
clean_df['Time'] = pd.to_datetime(clean_df['Time'])
ema_df['Time'] = pd.to_datetime(ema_df['Time'])
med_df['Time'] = pd.to_datetime(med_df['Time'])
raw_df['Time'] = pd.to_datetime(raw_df['Time'])
```

```python
# Assign colors for each dataframe
colors = ['blue', 'orange', 'green', 'red']
labels = ['Clean Data', 'EMA Data', 'Median Data', 'Raw Data']

# Create a figure and axis
plt.figure(figsize=(12, 8))

# Plot each dataframe
plt.plot(raw_df['Time'], raw_df['State'], color=colors[3], label=labels[3], linewidth=2)
plt.plot(ema_df['Time'], ema_df['State'], color=colors[1], label=labels[1], linewidth=2)
plt.plot(med_df['Time'], med_df['State'], color=colors[2], label=labels[2], linewidth=2)
plt.plot(clean_df['Time'], clean_df['State'], color=colors[0], label=labels[0], linewidth=2)

# Adding labels and title
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Comparison of Preproccesing methods')
plt.ylim(12, 45)
plt.legend()  # Show legend to differentiate between the dataframes
plt.grid(True)

# Show plot
plt.show()
```
